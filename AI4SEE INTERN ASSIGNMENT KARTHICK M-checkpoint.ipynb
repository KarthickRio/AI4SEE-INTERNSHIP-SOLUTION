```python
import matplotlib.pyplot as plt

#Set the number of rows and columns for subplots
num_features = len(psm_test_data.columns) - 1  
num_cols = 4  
num_rows = -(-num_features // num_cols) 

fig, axes = plt.subplots(num_rows, num_cols, figsize=(16, 4 * num_rows))

axes = axes.flatten()

# Plot each feature in a separate subplot
for i, column in enumerate(psm_test_data.columns[1:]):  
    axes[i].plot(psm_test_data[timestamp_column], psm_test_data[column], label=column)
    anomaly_regions = psm_test_label[psm_test_label['label'] == 1]
    axes[i].scatter(anomaly_regions[timestamp_column], psm_test_data[column][anomaly_regions.index], color='red', label='Anomaly Regions')
    axes[i].set_title(f'Time Series Plot of {column}')
    axes[i].set_xlabel('Timestamp')
    axes[i].set_ylabel('Value')
    axes[i].legend()
    axes[i].grid(True)


plt.tight_layout()
plt.show()
```

# d)Find out the variables which are the root cause for the anomaly

```python
# TO DO A ANAMOLY DETECTION USING MACHINE LEARNING MODEL
```

```python
from sklearn.ensemble import RandomForestClassifier

# Prepare data for modeling
X = psm_test_data.drop(columns=[ 'timestamp_(min)'])  # Features
y = psm_test_label['label']  # Target variable

# Train a random forest classifier
model = RandomForestClassifier()
model.fit(X, y)
```

```python
feature_importances = model.feature_importances_
print("Feature Importances:")
for feature, importance in zip(X.columns, feature_importances):
    print(f"{feature}: {importance}")
```
